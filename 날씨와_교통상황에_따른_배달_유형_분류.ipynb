```python
# prompt: Time_taken(min)열에서 "(min)"제거 후 숫자형 데이터로 변환

import pandas as pd

df = pd.read_csv('train.csv')

# Remove "(min)" from 'Time_taken(min)' column and convert to numeric
df['Time_taken(min)'] = df['Time_taken(min)'].str.replace('(min)', '', regex=False)
df['Time_taken(min)'] = pd.to_numeric(df['Time_taken(min)'], errors='coerce')

print(df.head())
```

```python
df['Weatherconditions'] = df['Weatherconditions'].str.replace('conditions ', '', regex=False)
print(df.head())
```

```python
# prompt: columns_drop을 사용하여 'ID' 열 제거
df = df.drop('ID', axis=1)

print(df.head())
```

```python
# prompt: Delivery_person_Age, Delivery_person_Ratings, Vehicle_condition 특성 Min Max scaling을 통한 정규화

from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler 객체 생성
scaler = MinMaxScaler()

# 'Delivery_person_Age', 'Delivery_person_Ratings', 'Vehicle_condition' 열 선택
columns_to_scale = ['Delivery_person_Age', 'Delivery_person_Ratings', 'Vehicle_condition']
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

print(df.head())
```

```python
# Drop 'Delivery_person_ID' column
df = df.drop('Delivery_person_ID', axis=1)

print(df.head())
```

```python
df = pd.get_dummies(df, columns=['Type_of_order', 'Type_of_vehicle', 'Festival', 'City','Weatherconditions','Road_traffic_density'])
```

```python
import pandas as pd

# NaN 값 처리 (Time_Order_picked에만 NaN 있음)
df['Time_Order_picked'] = df['Time_Order_picked'].fillna('00:00:00')

# datetime 변환 (체인 인덱싱 경고 방지)
df['Time_Orderd'] = pd.to_datetime(df['Time_Orderd'], format='%H:%M:%S', errors='coerce')
df['Time_Order_picked'] = pd.to_datetime(df['Time_Order_picked'], format='%H:%M:%S', errors='coerce')

# 변환 실패 값 확인
print("Time_Orderd NaT 개수:", df['Time_Orderd'].isna().sum())
print("Time_Order_picked NaT 개수:", df['Time_Order_picked'].isna().sum())
```

```python
# Step 1: NaT 행 확인
nat_rows = df[df['Time_Orderd'].isna()]
print("NaT가 발생한 행:", nat_rows)

# Step 2: NaT 값을 기본값으로 대체하거나 제거
df['Time_Orderd'] = pd.to_datetime(df['Time_Orderd'], format='%H:%M:%S', errors='coerce')

# Step 3: 변환 실패한 값 확인 및 처리
print("Time_Orderd NaT 개수:", df['Time_Orderd'].isna().sum())

# Step 4: 필요 시 NaT 행 제거
df = df[df['Time_Orderd'].notna()]
```

```python
#위도, 경도 삭제
df = df.drop(['Restaurant_latitude', 'Restaurant_longitude', 'Delivery_location_latitude', 'Delivery_location_longitude'], axis=1)
```

```python
# 결측치를 중앙값으로 채우기 (예시)
df['Delivery_person_Age'].fillna(df['Delivery_person_Age'].median(), inplace=True)
df['Delivery_person_Ratings'].fillna(df['Delivery_person_Ratings'].median(), inplace=True)
df['Vehicle_condition'].fillna(df['Vehicle_condition'].median(), inplace=True)

# 결측치 확인
print(df.isna().sum())
```

```python
print(df.dtypes)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 엘보우 메서드
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df[columns_to_scale])
    inertia.append(kmeans.inertia_)

# 결과 시각화
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
# 최적의 K 값으로 K-Means 모델 학습
optimal_k = 7  # 엘보우 메서드를 통해 결정된 값
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[columns_to_scale])
```

```python
# 군집화 후, 각 군집의 평균을 계산
df['Cluster'] = kmeans.labels_  # KMeans 레이블을 추가

# 군집별 평균 계산 (숫자 열만 선택하고 NaN 처리)
grouped = df.groupby('Cluster').mean(numeric_only=True)  # 숫자 데이터만 평균 계산

# 결과 출력
print(grouped)
```

```python
from sklearn.decomposition import PCA

# PCA로 차원 축소
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df[columns_to_scale])

# 시각화
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=df['Cluster'], cmap='viridis', alpha=0.6)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Clusters based on Weather and Traffic')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# 실루엣 점수 계산
silhouette_avg = silhouette_score(df[columns_to_scale], kmeans.labels_)
print(f"Silhouette Score: {silhouette_avg}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import silhouette_samples, silhouette_score

# 실루엣 점수 계산
silhouette_values = silhouette_samples(df[columns_to_scale], kmeans.labels_)

# 클러스터 개수
n_clusters = len(set(kmeans.labels_)) - (1 if -1 in kmeans.labels_ else 0)

# 실루엣 그래프 생성
fig, ax = plt.subplots(figsize=(10, 7))
y_lower = 10

for i in range(n_clusters):
    # 각 클러스터의 실루엣 점수 추출
    cluster_silhouette_values = silhouette_values[kmeans.labels_ == i]
    cluster_silhouette_values.sort()

    # 클러스터별 실루엣 그래프 그리기
    size_cluster_i = cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    ax.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        cluster_silhouette_values,
        alpha=0.7,
        label=f"Cluster {i+1}"
    )
    y_lower = y_upper + 10  # 다음 클러스터 시작점 조정

# 실루엣 점수 평균선 그리기
silhouette_avg = silhouette_score(df[columns_to_scale], kmeans.labels_)
ax.axvline(x=silhouette_avg, color="red", linestyle="--", label="Average Silhouette Score")

# 그래프 꾸미기
ax.set_title("Silhouette Plot for Each Cluster", fontsize=16)
ax.set_xlabel("Silhouette Coefficient", fontsize=12)
ax.set_ylabel("Cluster", fontsize=12)
ax.legend()
plt.tight_layout()
plt.show()
```
